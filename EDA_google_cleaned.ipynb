```python
cat_columns = ['Type']

height = len(cat_columns) * 3
aspect = 2


# Loop through each categorical column and create a plot

for col in cat_columns:
    sns.catplot(x=col, kind='count', data=cat_df, height=height, aspect=aspect)
    plt.xticks(rotation=45, ha='right') # Rotate the x-axis labels for better visibility
plt.show()
```

```python
cat_columns = ['Content Rating']
height = len(cat_columns)*3
aspect = 2

for col in cat_columns:
    sns.catplot(x = col, kind = 'count', data = cat_df, height = height, aspect = aspect)
    plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
# categorical columns
plt.figure(figsize = (15,15))
plt.suptitle("Univariate Analysis of categorical features", fontsize = 20, fontweight = 'bold', alpha = 0.8, y=1)
height = len(cat_fea)*10
aspect =2

cat_fea_1 = ['Type','Content Rating']
for i in range(0, len(cat_fea_1)):
    for col in cat_fea_1[i]:
        plt.subplot(2,2, i+1)
        sns.catplot(x = col, kind = 'count', data=cat_df, height = height, aspect =aspect)
        plt.xticks(rotation = 45, ha= 'right')
        sns.countplot()
```

```python
# categorical columns
plt.figure(fidsize = (15,15))
plt.suptitle("Univariate Analysis of Categorical Features", fontsize = 20, fontweight = 'bold', alpha =0.8, y=1)

cat_fea_1 = ["Type", "Content Rating"]
for i in range(0,len(cat_fea_1)):
    plt.subplot(2,2,i+1)
    sns.countplot()
```
